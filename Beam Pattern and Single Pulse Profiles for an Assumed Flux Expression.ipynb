```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import math 
%matplotlib inline
```

```python
a=1.07 #parameters for single pulse profile 1
b=1.28
z=6.1
w=np.arange(0,3*np.pi,0.01)
y=(1-(np.cos(a)*np.cos(b)+np.sin(a)*np.sin(b)*np.cos(6.1-w))**2)/2 #expression for the flux
plt.title('First Single Pulse Profile')
plt.xlabel('phase')
plt.ylabel('flux')
plt.plot(w,y)#plotting the flux w.r.t. phi for the first single pulse profile 
```

```python
aa=1.07 #parameters for single pulse profile 2
bb=2.1
zz=3.1
w=np.arange(0,3*np.pi,0.01)
ya=(1-(np.cos(aa)*np.cos(bb)+np.sin(aa)*np.sin(bb)*np.cos(w-zz))**2)/2
plt.title('Second Single Pulse Profile')
plt.xlabel('phase')
plt.ylabel('flux')
plt.plot(w,ya)
```

```python
#combined pulse profile
plt.title('Total Pulse Profile')
plt.xlabel('phase')
plt.ylabel('flux')
plt.plot(w,y+ya)
```

```python
#plotting the first SPP wrt Cos(-1)
g=np.cos(w-6.1)
plt.title('The First Single Pulse Profile wrt Cos(-1) ')
plt.xlabel('Cos(-1)')
plt.ylabel('flux')
plt.plot(g,y)
```

```python
#plotting the second SPP wrt Cos(-2)
f=np.cos(w-3.1)
plt.title('The Second Single Pulse Profile wrt Cos(-2)')
plt.xlabel('Cos(-2)')
plt.ylabel('flux')
plt.plot(f,ya)
```

```python
m=np.arange(0,np.pi,0.01)
flux=((np.sin(m))**2)/2
plt.title('Beam Pattern')
plt.xlabel('')
plt.ylabel('flux')
plt.plot(m,flux)
```
